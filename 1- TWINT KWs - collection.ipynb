```
import pandas as pd
import json
import datetime
today = datetime.datetime.today().strftime("%Y-%m-%d")
init_args_dict = dict(limit=90000)

def generate_path(kw: str):
  """
  Args:
    kw (str): keyword
  """
  if '"' not in kw:
    path = "s3://REDACTED/orbl/twint/amazon/dt={0}/{1}.csv".format(today, "_".join(kw.split(" ")))
  else:
    kw = "exact_" + kw.strip('"')
    path = "s3://REDACTED/orbl/twint/amazon/dt={0}/{1}.csv".format(today, "_".join(kw.split(" ")))
  return path

paths = []
kws = ["amazon website", "amazon site", '"amazon website"', '"amazon site"', 'amazon app', 'amazon application', '"amazon app"', '"amazon application"']
for kw in kws:
  request_args = dict(kw=kw)
  results = Twint(init_args_dict).get_tweets_by_kw(request_args)
  print(results)
  results_df = pd.DataFrame(json.loads(results).get("Items"))
  results_df["query"] = kw
  results_df["type"] = "website" if "site" in kw else "application"
  path = generate_path(kw)
  paths.append(path)
  results_df.to_csv(path, index = False)
```
